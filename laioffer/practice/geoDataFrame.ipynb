```python
_ = libpysal.examples.load_example('SanFran Crime')
pth = libpysal.examples.get_path('SFCrime_blocks.shp')
df = read_file(pth)
df.head()
```

```python
# %matplotlib inline
# ax = df.plot(column='CarTheft', scheme='QUANTILES', k=4, \
#              cmap='BuPu', legend=True,
#              legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)})



# Just a figure and one subplot
fig, ax = plt.subplots(2,2)
df.plot(column='Vandalism', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=False,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},\
            ax = ax[0][0])
df.plot(column='Robbery', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=False,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},\
            ax = ax[0][1])
df.plot(column='Drugs', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=False,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},\
            ax = ax[1][0])
df.plot(column='CarTheft', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=False,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},\
            ax = ax[1][1])

            
fig.set_size_inches(12,12)
ax[0][0].set_title('Vandalism')
ax[0][1].set_title('Robbery')
ax[1][0].set_title('Drugs')
ax[1][0].set_title('CarTheft')
```
