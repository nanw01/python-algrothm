```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""


class Solution:
    """
    @param root: An object of TreeNode, denote the root of the binary tree.
    This method will be invoked first, you should design your own algorithm 
    to serialize a binary tree which denote by a root node to a string which
    can be easily deserialized by your own "deserialize" method later.
    """
    def serialize(self, root):
        if root is None:
            return '{}'
        nodes = []
        queue = collections.deque()
        
        # 宽度优先搜索，获得BFS序
        queue.append(root)
        while queue:
            node = queue.popleft()
            nodes.append(node)
            if node is None:
                continue
            queue.append(node.left)
            queue.append(node.right)
        
        while nodes[-1] is None:
            nodes.pop()
        
        # 将BFS序序列化
        result = '{'
        for i in range(len(nodes)):
            if i > 0:
                result += ','
            if nodes[i] is None:
                result += '#'
            else:
                result += str(nodes[i].val)
        result += '}'
        
        return result

    """
    @param data: A string serialized by your serialize method.
    This method will be invoked second, the argument data is what exactly
    you serialized at method "serialize", that means the data is not given by
    system, it's given by your own serialize method. So the format of data is
    designed by yourself, and deserialize it here as you serialize it in 
    "serialize" method.
    """
    def deserialize(self, data):
        if data == '{}':
            return None
        bfs_order = data[1:-1].split(',')
        n = len(bfs_order)
        root = self.get_treenode(bfs_order[0])
        queue = collections.deque()
        pointer = 1
        
        # 分层进行宽度优先搜索
        # 获得该层的节点个数level_size，并将接下来level_size * 2个值连接到节点上
        queue.append(root)
        while queue:
            if pointer < n:
                level_size = len(queue)
                for _ in range(level_size):
                    node = queue.popleft()
                    
                    node.left = self.get_treenode(bfs_order[pointer])
                    if node.left is not None:
                        queue.append(node.left)
                    pointer += 1
                    if pointer == n:
                        break
                    
                    node.right = self.get_treenode(bfs_order[pointer])
                    if node.right is not None:
                        queue.append(node.right)
                    pointer += 1
                    if pointer == n:
                        break
            else:
                break
        return root
    
    def get_treenode(self, s):
        if s == '#':
            return None
        else:
            return TreeNode(int(s))
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        ret = []
        nums.sort()
        for i in range(len(nums)-2):
            if nums[i] >0: break;
            if i > 0 and nums[i] == nums[i - 1]: continue
            l = i+1
            r = len(nums)-1

            while l<r:
                temp = nums[l]+nums[r]
                if temp+nums[i] == 0:
                    ret.append([nums[i],nums[l], nums[r]])
                    l+=1
                    r-=1
                    while l<r and nums[l] == nums[l-1]: l += 1
                    while l<r and nums[r] == nums[r+1]: r -= 1
                    
                elif temp+nums[i] < 0:
                    l+=1
                    while l<r and nums[l] == nums[l-1]: l += 1
                elif temp+nums[i] >0:
                    r-=1
                    while l<r and nums[r] == nums[r+1]: r -= 1
                else:
                    pass
        return ret
```

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:

        n = len(nums)
        output = [[]]
        
        for num in nums:
            output += [curr + [num] for curr in output]
        
        return output
```

```python


from heapq import *
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:

        heap = []
        heapify(heap)
        for num in nums:
            heappush(heap, num)
            if len(heap) > k:
                heappop(heap)
        return heap[0]

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:

        return sorted(nums, reverse = True)[k-1]

```
